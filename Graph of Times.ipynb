This .ipynb was used to generate plots for the times it took to run the algorithms from the data generated in the previous notebook.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Plot the time it takes to run each algorithm on classical backend vs. number of qubits / states in circuit
qubit_num= [2**n for n in range(2, 11)]
time_grovers = np.array([1.362600564956665, 0.22480106353759766, 0.5269198417663574, 0.8145365715026855, 2.031658887863159, 6.317428112030029, 11.401068210601807, 36.35035490989685, 248.46139764785767])
time_grovers /= 1000
time_classical= [0.0001234709999948791, 0.024240263000024243, 0.011649404999957369, 0.013785587000029409, 0.007511383999997179, 0.011057438000023012, 0.026812333999998827, 0.014148534000014479, 0.04724741100005758]
```

```python
plt.plot(qubit_num, time_grovers, label = "Grovers")
plt.plot(qubit_num, time_classical, label = "Classical")
plt.xlabel('Number of States')
plt.ylabel("Runtime (s)")
plt.title("Runtime for Quantum vs. Classical Search Algorithm On Fake Backend")
plt.legend()
plt.show()
```
