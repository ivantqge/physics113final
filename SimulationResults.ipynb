This notebook is used to visualize and plot the simualtion results from the fake backend on different lengths of qubit inputs.

Results from 3 qubit simulations

```python
import matplotlib.pyplot as plt
import pandas as pd

# Results from backend simulation, proportion of measurements that were the marked state
results = {
    'Vigo': 0.5998046875,
    'Toronto': 0.1150390625,
    'Tokyo': 0.4529296875,
    'Casa': 0.54375,
    'Guada': 0.5759765625,
    'Melbo': 0.3521484375,
    'Kyiv': 0.7068359375,
    'Osaka': 0.5333984375,
    'Qasm': 0.9462890625
}

# Number of qubits for each quantum computer
num_of_qubits = {
    'Vigo': 5,
    'Toronto': 27,
    'Tokyo': 20,
    'Casa': 7,
    'Guada': 16,
    'Melbo': 17,
    'Kyiv': 127,
    'Osaka' : 127,
    'Qasm' : 0
}

# Pandas dataframe
data = pd.DataFrame({
    'Backend': list(results.keys()),
    'Probability': [results[backend] for backend in results.keys()],
    'Qubits': [num_of_qubits[backend] for backend in results.keys()]
})

# Sort by the number of qubits
data = data.sort_values(by='Qubits')

# Plotting
plt.bar(data['Backend'], data['Probability'], color='blue')
plt.xlabel('Backend')
plt.ylabel('Proportion of Marked States Measured')
plt.title('Probability of Measuring Marked State for Different Backends')
plt.grid(axis='y')

plt.show()
```
